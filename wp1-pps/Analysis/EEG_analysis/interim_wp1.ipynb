```python
import mne
import matplotlib.pyplot as plt
import numpy as np
import glob
from mne.channels import read_custom_montage
from niseq import sequential_cluster_test_1samp
```

```python
montage = read_custom_montage('//kfs/fileserver/Lab/Equipment/EEG/Montage Files-128NOREF/CACS-64_NO_REF.bvef')
eog_map = {
    'HEOG L': 'eog',
    'HEOG R': 'eog',
    'VEOG L': 'eog',
    'VEOG U': 'eog',
    'E': 'eog'
}
```

```python
conditions = ['PPS_hi', 'EPS_hi', 'PPS_lo', 'EPS_lo']
data_dir = '//kfs/fileserver/Projektdaten/CherISH/data/wp-1/EEG/09_fifData/'
data_files_pps_hi = glob.glob(data_dir + '*pps_hi-ave.fif')
data_files_eps_hi  = glob.glob(data_dir + '*eps_hi-ave.fif')
data_files_pps_lo = glob.glob(data_dir + '*pps_lo-ave.fif')
data_files_eps_lo  = glob.glob(data_dir + '*eps_lo-ave.fif')
```

```python

evokeds_pps_hi = {}

evokeds_pps_hi = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_pps_hi
    ]

evokeds_eps_hi = {}

evokeds_eps_hi = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_eps_hi
    ]
```

```python
evokeds_pps_lo = {}

evokeds_pps_lo = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_pps_lo
    ]

evokeds_eps_lo = {}

evokeds_eps_lo = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_eps_lo
    ]
```

```python
## Cluster based permutation test
diff_EPS_PPS = [mne.combine_evoked([evokeds_eps_hi[subj],
                                       evokeds_eps_lo[subj],
                                        evokeds_pps_hi[subj],
                                       evokeds_pps_lo[subj]
                                        ],
                                 weights=[1,1,-1,-1]
                                 )
              for subj in range(len(data_files_eps_hi))]
```

```python
roi = ['Cz']
contrast = 'EPS-PPS difference'
mne.viz.plot_compare_evokeds({contrast:diff_EPS_PPS}, combine='mean',
                            legend=None,
                            picks=roi, show_sensors='upper right',
                            title=contrast
                            )
plt.show()
```

```python
## Interim analysis pooling high and low intensity
X=np.stack([evo.get_data(picks='eeg') for evo in diff_EPS_PPS], axis =0)
X=np.transpose(X, (0,2,1))

adj,_=mne.channels.find_ch_adjacency(diff_EPS_PPS[0].info, "eeg")


#n=X.shape[0]
n=[10,22]
looks, ps, adj_alphas, spending = sequential_cluster_test_1samp(X,
                                                                look_times=n,
                                                                n_max=30,
                                                                alpha=0.05,
                                                                tail=0,
                                                                verbose=True,
                                                                n_permutations = 50,
                                                                adjacency=adj,
                                                                out_type = 'indices')

#t_map, clusts, clust_ps, H0 = looks[n]
```
