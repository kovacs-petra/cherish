```python
import mne
import matplotlib.pyplot as plt
import numpy as np
import glob
from mne.channels import read_custom_montage
```

```python
montage = read_custom_montage('//kfs/fileserver/Lab/Equipment/EEG/Montage Files-128NOREF/CACS-64_NO_REF.bvef')
eog_map = {
    'HEOG L': 'eog',
    'HEOG R': 'eog',
    'VEOG L': 'eog',
    'VEOG U': 'eog',
    'E': 'eog'
}
```

```python
conditions = ['loom_hi', 'rec_hi', 'loom_lo', 'rec_lo']
data_dir = '//kfs/fileserver/Projektdaten/CherISH/data/wp-1/EEG/09_fifData/'
data_files_loom_hi = glob.glob(data_dir + '*loom_hi-ave.fif')
data_files_rec_hi  = glob.glob(data_dir + '*rec_hi-ave.fif')
data_files_loom_lo = glob.glob(data_dir + '*loom_lo-ave.fif')
data_files_rec_lo  = glob.glob(data_dir + '*rec_lo-ave.fif')
data_files_pps_hi = glob.glob(data_dir + '*pps_hi-ave.fif')
data_files_pps_lo = glob.glob(data_dir + '*pps_lo-ave.fif')
```

```python
data_files_loom_all = glob.glob(data_dir + '*loom_*-ave.fif')
evokeds_loom_all = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_loom_all
    ]
```

```python
evokeds_loom_hi = {}

evokeds_loom_hi = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_loom_hi
    ]

evokeds_loom_hi

evokeds_rec_hi = {}

evokeds_rec_hi = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_rec_hi
    ]

evokeds_rec_hi
```

```python
evokeds_loom_lo = {}

evokeds_loom_lo = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_loom_lo
    ]

evokeds_loom_lo

evokeds_rec_lo = {}

evokeds_rec_lo = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_rec_lo
    ]

evokeds_rec_lo
```

```python
## Cluster visualization
diff_rec_loom = [mne.combine_evoked(
    [evokeds_rec_hi[subj], evokeds_rec_lo[subj], evokeds_loom_hi[subj], evokeds_loom_lo[subj]], weights=[1, 1, -1, -1])
              for subj in range(len(data_files_rec_hi))]

roi = ['Cz']
contrast = 'Receding-looming difference'
mne.viz.plot_compare_evokeds({contrast:diff_rec_loom}, combine='mean',
                            legend=None,
                            picks=roi, show_sensors='upper right',
                            title=contrast
                            )
plt.show()
```

```python
## Cluster-based permutation test
X=np.stack([evo.get_data(picks='eeg') for evo in diff_rec_loom], axis =0)
X=np.transpose(X, (0,2,1))

adj,_=mne.channels.find_ch_adjacency(diff_rec_loom[0].info, "eeg")

F_obs, clusters, cluster_pv, H0 = mne.stats.permutation_cluster_1samp_test(X,
                                                                n_permutations=1000,
                                                                tail=0,
                                                                verbose=True,
                                                                adjacency=adj)
```

```python
alpha = 0.05
sig_cluster_inds = np.where(cluster_pv < alpha)[0]
sig_cluster_inds
min(cluster_pv)

```

```python
## Cluster-based permutation test
X=np.stack([evo.get_data(picks='eeg') for evo in diff_rec_pps], axis =0)
X=np.transpose(X, (0,2,1))

adj,_=mne.channels.find_ch_adjacency(diff_rec_pps[0].info, "eeg")

F_obs, clusters, cluster_pv, H0 = mne.stats.permutation_cluster_1samp_test(X,
                                                                n_permutations=500,
                                                                tail=0,
                                                                verbose=True,
                                                                adjacency=adj)
```
