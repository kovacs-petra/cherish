```python
import mne
import matplotlib.pyplot as plt
import numpy as np
import glob
from mne.channels import read_custom_montage
from mne.datasets import sample
from mne.minimum_norm import apply_inverse, make_inverse_operator
```

```python
# Define color scheme and linestyles for plotting
loom_color = "#dc3220"
rec_color = "#1A85FF"
pps_color = "#4B0092"
eps_color = "#1AFF1A"

loom_linestyle = "solid"
rec_linestyle = "solid"
pps_linestyle = "dashed"
eps_linestyle = "dashed"
```

```python
montage = read_custom_montage('//kfs/fileserver/Lab/Equipment/EEG/Montage Files-128NOREF/CACS-64_NO_REF.bvef')
eog_map = {
    'HEOG L': 'eog',
    'HEOG R': 'eog',
    'VEOG L': 'eog',
    'VEOG U': 'eog',
    'E': 'eog'
}
```

```python
conditions = ['loom', 'rec', 'pps', 'eps']
data_dir = '//kfs/fileserver/Projektdaten/CherISH/data/wp-1/EEG/09_fifData/CNV/'
data_files_loom = glob.glob(data_dir + '*loom-ave.fif')
data_files_rec  = glob.glob(data_dir + '*rec-ave.fif')
data_files_pps = glob.glob(data_dir + '*pps-ave.fif')
data_files_eps = glob.glob(data_dir + '*eps-ave.fif')
```

```python
# Read evokeds for each condition
evokeds_loom = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_loom
    ]

evokeds_rec = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_rec
    ]

evokeds_pps = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_pps
    ]

evokeds_eps = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_eps
    ]
```

```python
# Add condition names
for subj in range(len(data_files_loom)):
    evokeds_loom[subj].comment = "Looming"
    evokeds_loom[subj].data *= 1e-6
    evokeds_rec[subj].comment = "Receding"
    evokeds_rec[subj].data *= 1e-6
    evokeds_pps[subj].comment = "Rotating in PPS"
    evokeds_pps[subj].data *= 1e-6
    evokeds_eps[subj].comment = "Rotating in EPS"
    evokeds_eps[subj].data *= 1e-6
```

```python
## Cluster visualization Loom-Rec
diff_loom_rec = [mne.combine_evoked(
    [evokeds_rec[subj], evokeds_loom[subj]], weights=[-1, 1])
              for subj in range(len(data_files_rec))]

# for subj in range(len(data_files_loom)):
#     evokeds_loom[subj].comment = "Looming"
#     evokeds_rec[subj].comment = "Receding"
#     evokeds_pps[subj].comment = "Rotating in PPS"
#     evokeds_eps[subj].comment = "Rotating in EPS"

roi = ['Cz']
contrast = 'Looming-receding difference'
mne.viz.plot_compare_evokeds({contrast:diff_loom_rec}, combine='mean',
                            legend=None,
                            picks=roi, show_sensors='upper right',
                            title=contrast
                            )
plt.show()
```

```python
## Cluster visualization EPS-PPS
diff_PPS_EPS = [mne.combine_evoked(
    [evokeds_eps[subj], evokeds_pps[subj]], weights=[-1, 1])
              for subj in range(len(data_files_eps))]

roi = ['Cz']
contrast = 'PPS-EPS difference'
mne.viz.plot_compare_evokeds({contrast:diff_PPS_EPS}, combine='mean',
                            legend=None,
                            picks=roi, show_sensors='upper right',
                            title=contrast
                            )
plt.show()
```

```python
## Cluster-based permutation test Rec vs Loom
X=np.stack([evo.get_data(picks='eeg') for evo in diff_loom_rec], axis =0)
X=np.transpose(X, (0,2,1))

adj,_=mne.channels.find_ch_adjacency(diff_loom_rec[0].info, "eeg")

F_obs, clusters, cluster_pv, H0 = mne.stats.permutation_cluster_1samp_test(X,
                                                                n_permutations=1000,
                                                                tail=0,
                                                                verbose=True,
                                                                adjacency=adj)

alpha = 0.05
sig_cluster_inds = np.where(cluster_pv < alpha)[0]
```

```python
F_obs = np.transpose(F_obs)
print(F_obs.shape)
```

```python
# Plot channel-time img loom-rec
plotting_diff_loom_rec = diff_loom_rec[0].copy()

# Initialize a mask for significant cluster samples
sig_mask = np.zeros(F_obs.shape, dtype=bool)

for i_clu in sig_cluster_inds:
    ch_inds,time_inds = clusters[i_clu]
    sig_mask[np.ix_(time_inds,ch_inds)] = True

np.unique(time_inds)
```

```python

fig, ax = plt.subplots(figsize=(10, 6))

im = ax.imshow(
    F_obs,
    aspect='auto',
    origin='lower',
    extent=[
        plotting_diff_loom_rec.times[0],
        plotting_diff_loom_rec.times[-1],
        0,
        F_obs.shape[0]
    ],
    cmap='RdBu_r'
)

# Overlay significant points
ax.contour(
    sig_mask,
    colors='black',
    linewidths=0.5,
    origin='lower',
    extent=[
        plotting_diff_loom_rec.times[0],
        plotting_diff_loom_rec.times[-1],
        0,
        F_obs.shape[0]
    ]
)

ax.set_xlabel("Time (s)")
ax.set_ylabel("Channel index")
ax.set_title("Looming-Receding")
fig.colorbar(im, ax=ax, label="t-value")

plt.show()
```

```python
# Plot time course with topography rec-loom
evokeds_dict_loom_rec = {
    "Looming": evokeds_loom,
    "Receding": evokeds_rec,
}

for i_clu in sig_cluster_inds:
    time_inds,ch_inds = clusters[i_clu]
    times = diff_loom_rec[0].times[time_inds]

    # Convert indices to channel names
    ch_names = np.array(diff_loom_rec[0].ch_names)[np.unique(ch_inds)]

    # Keep EEG channels only
    ch_names_eeg = [
        ch for ch in ch_names
        if diff_loom_rec[0].get_channel_types(picks=[ch])[0] == "eeg"
    ]

    # Average t-values over time inside the cluster
    t_map = F_obs[:, time_inds].mean(axis=1)

    fig, ax = plt.subplots(1, 2, figsize=(10, 4))

    # Topography of the cluster
    mask = np.zeros((len(t_map), 1), dtype=bool)
    mask[ch_inds, :] = True

    im,cm = mne.viz.plot_topomap(
        t_map,
        diff_loom_rec[0].info,
        axes=ax[0],
        mask=mask,
        show=False,
        contours=0,
        cmap="Greys"
    )

    ax_x_start = 0.08
    ax_x_width = 0.04
    ax_y_start = 0.1
    ax_y_height = 0.2
    cbar_ax = fig.add_axes([ax_x_start, ax_y_start, ax_x_width, ax_y_height])
    clb = fig.colorbar(im, cax=cbar_ax)
    clb.ax.set_title("t") # title on top of colorbar

    ax[0].set_title(
        f"{times[0]*1e3:.0f}{times[-1]*1e3:.0f} ms\n"
        f"p = {cluster_pv[i_clu]:.3f}"
    )

    # Mean ERPs on the cluster's channels
    ax[1].hlines(
            y=0.05,
            xmin=times[0],
            xmax=times[-1],
            transform=ax[1].get_xaxis_transform(),
            colors="k",
            linewidth=4
            )

    mne.viz.plot_compare_evokeds(evokeds_dict_loom_rec, combine='mean',
                            legend=True,
                            axes=ax[1],
                            picks=ch_names_eeg,
                            show_sensors='lower left',
                            vlines=[],
                            title="ERPs on cluster channels",
                            cmap=None,
                            colors=[loom_color,rec_color],
                            linestyles=[loom_linestyle,rec_linestyle],
                            truncate_xaxis=False,
                            truncate_yaxis=False
                            )

    plt.tight_layout()
    plt.show()
```
