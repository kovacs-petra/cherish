```python
import mne
import matplotlib.pyplot as plt
import numpy as np
import glob
from mne.channels import read_custom_montage
```

```python
# Define color scheme and linestyles for plotting
pps_color = "#4B0092"
eps_color = "#1AFF1A"

pps_linestyle = "dashed"
eps_linestyle = "dashed"
```

```python
montage = read_custom_montage('//kfs/fileserver/Lab/Equipment/EEG/Montage Files-128NOREF/CACS-64_NO_REF.bvef')
eog_map = {
    'HEOG L': 'eog',
    'HEOG R': 'eog',
    'VEOG L': 'eog',
    'VEOG U': 'eog',
    'E': 'eog'
}
```

```python
data_dir = '//kfs/fileserver/Projektdaten/CherISH/data/wp-1/EEG/09_fifData/PPSlo-EPShi/'
data_files_pps_lo = glob.glob(data_dir + '*pps_lo-ave.fif')
data_files_pps_hi = glob.glob(data_dir + '*pps_hi-ave.fif')
data_files_eps_hi = glob.glob(data_dir + '*eps_hi-ave.fif')
data_files_eps_lo = glob.glob(data_dir + '*eps_lo-ave.fif')
```

```python
# Read evokeds for each condition
evokeds_pps_lo = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_pps_lo
    ]

evokeds_eps_hi = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_eps_hi
    ]

evokeds_pps_hi = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_pps_hi
    ]

evokeds_eps_lo = [
        mne.read_evokeds(d)[0]
        .set_montage(montage, on_missing='ignore')
        .set_channel_types(eog_map)
        for d in data_files_eps_lo
    ]
```

```python
# Add condition names, calculate difference waves
for subj in range(len(data_files_pps_hi)):
    evokeds_pps_lo[subj].comment = "Rotating in PPS, low int"
    evokeds_pps_lo[subj].data *= 1e-6
    evokeds_eps_hi[subj].comment = "Rotating in EPS, high int"
    evokeds_eps_hi[subj].data *= 1e-6

    evokeds_pps_hi[subj].comment = "Rotating in PPS, high int"
    evokeds_pps_hi[subj].data *= 1e-6
    evokeds_eps_lo[subj].comment = "Rotating in EPS, low int"
    evokeds_eps_lo[subj].data *= 1e-6

diff_PPSlo_EPShi = [mne.combine_evoked(
    [evokeds_eps_hi[subj], evokeds_pps_lo[subj]], weights=[-1, 1])
              for subj in range(len(data_files_eps_hi))]

diff_PPS_EPS_overall = [mne.combine_evoked(
    [evokeds_eps_hi[subj],evokeds_eps_lo[subj],evokeds_pps_hi[subj],evokeds_pps_lo[subj]], weights=[-1,-1,1,1])
              for subj in range(len(data_files_eps_lo))]
```

```python
## Cluster-based permutation test EPS vs PPS
Y=np.stack([evo.get_data(picks='eeg') for evo in diff_PPSlo_EPShi], axis =0)
Y=np.transpose(Y, (0,2,1))

adj,_=mne.channels.find_ch_adjacency(diff_PPSlo_EPShi[0].info, "eeg")

F_obs2, clusters2, cluster_pv2, H02 = mne.stats.permutation_cluster_1samp_test(Y,
                                                                n_permutations=1000,
                                                                tail=0,
                                                                verbose=True,
                                                                adjacency=adj)
```

```python
alpha = 0.05
sig_cluster_inds2 = np.where(cluster_pv2 < alpha)[0]
sig_cluster_inds2
```

```python
F_obs2 = np.transpose(F_obs2)
print(F_obs2.shape)
```

```python
# Plot channel-time img EPS-PPS
plotting_diff_pps_eps = diff_PPS_EPS[0].copy()

# Initialize a mask for significant cluster samples
sig_mask2 = np.zeros(F_obs2.shape, dtype=bool)

for i_clu in sig_cluster_inds2:
    ch_inds2,time_inds2 = clusters2[i_clu]
    sig_mask2[np.ix_(time_inds2,ch_inds2)] = True

fig, ax = plt.subplots(figsize=(10, 6))

im = ax.imshow(
    F_obs2,
    aspect='auto',
    origin='lower',
    extent=[
        plotting_diff_pps_eps.times[0],
        plotting_diff_pps_eps.times[-1],
        0,
        F_obs2.shape[0]
    ],
    cmap='RdBu_r'
)

# Overlay significant points
ax.contour(
    sig_mask2,
    colors='black',
    linewidths=0.5,
    origin='lower',
    extent=[
        plotting_diff_pps_eps.times[0],
        plotting_diff_pps_eps.times[-1],
        0,
        F_obs2.shape[0]
    ]
)

ax.set_xlabel("Time (s)")
ax.set_ylabel("Channel index")
ax.set_title("PPS-EPS")
fig.colorbar(im, ax=ax, label="t-value")

plt.show()
```

```python
# Plot time course with topography EPS-PPS
evokeds_dict_pps_eps = {
    "Rotating in PPS, low int": evokeds_pps_lo,
    "Rotating in EPS, high int": evokeds_eps_hi,
}
for i_clu in sig_cluster_inds2:
    time_inds2,ch_inds2 = clusters2[i_clu]
    times2 = diff_PPS_EPS[0].times[time_inds2]

    # Convert indices to channel names
    ch_names2 = np.array(diff_PPS_EPS[0].ch_names)[np.unique(ch_inds2)]

    # Keep EEG channels only
    ch_names_eeg2 = [
        ch for ch in ch_names2
        if diff_PPS_EPS[0].get_channel_types(picks=[ch])[0] == "eeg"
    ]

    # Average t-values over time inside the cluster
    t_map2 = F_obs2[:, time_inds2].mean(axis=1)

    fig, ax = plt.subplots(1, 2, figsize=(10, 4))

    # Topography of the cluster
    mask2 = np.zeros((len(t_map2), 1), dtype=bool)
    mask2[ch_inds2, :] = True

    im,cm = mne.viz.plot_topomap(
        t_map2,
        diff_PPS_EPS[0].info,
        axes=ax[0],
        mask=mask2,
        show=False,
        contours=0,
        cmap="Greys"
    )

    ax_x_start = 0.08
    ax_x_width = 0.04
    ax_y_start = 0.1
    ax_y_height = 0.2
    cbar_ax = fig.add_axes([ax_x_start, ax_y_start, ax_x_width, ax_y_height])
    clb = fig.colorbar(im, cax=cbar_ax)
    clb.ax.set_title("t") # title on top of colorbar

    ax[0].set_title(
        f"{times2[0]*1e3:.0f}{times2[-1]*1e3:.0f} ms\n"
        f"p = {cluster_pv2[i_clu]:.3f}"
    )

    # Mean ERPs on the cluster's channels
    ax[1].hlines(
            y=0.05,
            xmin=times2[0],
            xmax=times2[-1],
            transform=ax[1].get_xaxis_transform(),
            colors="k",
            linewidth=4
            )

    mne.viz.plot_compare_evokeds(evokeds_dict_pps_eps, combine='mean',
                            legend=True,
                            axes=ax[1],
                            picks=ch_names_eeg2,
                            show_sensors='lower left',
                            vlines=[],
                            title="ERPs on cluster channels",
                            cmap=None,
                            colors=[pps_color,eps_color],
                            linestyles=[pps_linestyle,eps_linestyle],
                            truncate_xaxis=False,
                            truncate_yaxis=False
                            )


    plt.tight_layout()
    plt.show()

```

```python
roi = ['Cz']
contrast = 'PPS, low int - EPS, high int difference'
mne.viz.plot_compare_evokeds({contrast:diff_PPS_EPS}, combine='mean',
                            legend=None,
                            picks=roi, show_sensors='upper right',
                            title=contrast,
                            truncate_xaxis=False,
                            truncate_yaxis=False
                            )
plt.show()
```
